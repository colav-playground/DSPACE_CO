```python
from colombia_config import endpoints
from colombia_config2 import endpoints as endpoints2
from colombia_config3 import endpoints as endpoints3
from colombia_config4 import endpoints as endpoints4
import requests
import ssl
import urllib
import socket
from bs4 import BeautifulSoup
```

```python
#setting https and timeout support for urllib
ssl_context = ssl._create_unverified_context()

# Configurar un opener global con el contexto SSL
opener = urllib.request.build_opener(urllib.request.HTTPSHandler(context=ssl_context))
urllib.request.install_opener(opener)

socket.setdefaulttimeout(240)
```

```python
def get_ror_name(ror_id):
    url = "https://api.ror.org/v1/organizations/"+ror_id.split("/")[-1]
    req = requests.get(url)
    data=req.json()
    return data["name"]
```
